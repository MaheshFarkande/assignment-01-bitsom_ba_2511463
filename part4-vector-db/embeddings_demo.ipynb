```python
from sentence_transformers import SentenceTransformer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    # Cricket
    'The batsman scored a century with elegant cover drives.',
    'The fast bowler generated swing under cloudy conditions.',
    'The cricket match went into a tense final over.',

    # Cooking
    'The chef simmered the curry slowly to enhance the flavors.',
    'Baking bread requires precise measurements and patience.',
    'Fresh herbs can completely transform a home-cooked meal.',

    # Cybersecurity
    'Multi-factor authentication helps protect user accounts.',
    'The company suffered a data breach due to a phishing attack.',
    'Encrypting sensitive data is critical for cybersecurity.',
    'Regular patching reduces vulnerabilities in software systems.'
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings)
plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, xticklabels=False, yticklabels=False, cmap='viridis')
plt.title('Cosine Similarity Heatmap')
plt.show()
```

```python
query = 'The bowler took three wickets in one over'
query_embedding = model.encode([query])
scores = cosine_similarity(query_embedding, embeddings)[0]
top_indices = scores.argsort()[-2:][::-1]
for idx in top_indices:
    print(f'Sentence: {sentences[idx]}')
    print(f'Similarity Score: {scores[idx]:.4f}\n')
```
